```python
from collections import defaultdict
from bitarray import bitarray
import random
import math
```

# ***General values***

```python
prop_a = 0.4
prop_b = 0.3
prop_c = 0.2
prop_d = 0.1
```

# ***Generate a random sequence of symbols a, b, c, d, such that P(a)=0.4, P(b)=03, P(c) = 0.2, P(d)=0.1***

```python
def generate_sequence(sequence_size, sequence):
  #generate a, b, c, d depends on their probabilites
  sequence += list(random.choice('a') for _ in range(int(sequence_size * prop_a)) )
  sequence += list(random.choice('b') for _ in range(int(sequence_size * prop_b)) )
  sequence += list(random.choice('c') for _ in range(int(sequence_size * prop_c)) )
  sequence += list(random.choice('d') for _ in range(int(sequence_size * prop_d)) )

  # shuffle all characters
  random.SystemRandom().shuffle(sequence)
  sequence = ''.join(sequence)   #convert list to string

  print("The random sequnce generated with size = ", sequence_size, "symbols is ",sequence)
```

# ***Find Source Entropy H***

```python
def find_entropy(sequence):
  Source_entropy = 0

  freq_lib = defaultdict(int)          #generate a default library
  for ch in sequence:                  #count each letter and record into the frequency library
    freq_lib[ch]+=1

  Source_entropy =  freq_lib['a']*(prop_a * math.log2(1/prop_a)) + freq_lib['b']*(prop_b * math.log2(1/prop_b)) + freq_lib['c']*(prop_c * math.log2(1/prop_c)) + freq_lib['d']*(prop_d * math.log2(1/prop_d))

  print("Entropy of the alphabet = ", Source_entropy, "bit/ symbol")
```

# ***Parse the data and assign a number to each phrase***

```python
def parsing_and_findEncodedPackets(dictionary_contents, encoded_packets):
  ch = 0

  dictionary_contents_table = []

  p = sequence[ch]                            #take the first element in the sequence

  while(ch != (len(sequence) - 1)):
  
    if p in dictionary_contents_table:        # if p exist in the sequnce
      c = sequence[ch+1]
      ch +=1
      encoded_num = dictionary_contents_table.index(p) + 1
      p = p + c
    if not p in dictionary_contents_table:    #if p not on the table
      dictionary_contents_table.append(p)    
      if len(p) == 1:
        encoded_packets[p] = 0
      else:
        encoded_packets[p] = encoded_num 
      if ch != (len(sequence) - 1):
        ch += 1  
        p = sequence[ch]  

  for i in range(len(dictionary_contents_table)):
    dictionary_contents[i+1] = dictionary_contents_table[i]
    
  print("Parsing data with number assigned each: \n", dictionary_contents)

  print("Encoded Packets generated: \n", encoded_packets)
```

# ***Find the number of binary digits needed to encode the sequence***

```python
def find_NB(dictionary_contents, encoded_packets):
  LZ_compression_bits_needed = len(dictionary_contents) * (8 + len(format(  max(list(encoded_packets.values())  ), 'b'))   )

  print("With LZ compression, the number of bits needed is ((NB)):", LZ_compression_bits_needed)

  return( LZ_compression_bits_needed)
```

# ***Number of bits per phrase assuming that the tail of each codeword is in ASCII format (8 bits).***

```python
def bits_per_phrase(encoded_packets):
  bits_per_phrase = 8 + len(format(  max(list(encoded_packets.values())  ), 'b'))

  print("number of bits per phrase assuming that the tail of each codeword is in ASCII format (8 bits) = ", bits_per_phrase)
```

# ***Find the compression ratio relative to the ASCII code***

```python
def commp_ratio(NB_avg, sequence_size):
  commpression_ratio = (NB_avg/ (8*sequence_size)) * 100

  print("The compression ratio relative to the ASCII code = ", commpression_ratio , "%")
```

# ***Repeat parts 4 and 5 five times and find the average value of (NB)***
# ***Parsing the data, find NB and bits per symbol***

```python
sequence_size = 20
dictionary_contents = {}                    #Parsing data
encoded_packets = {}                        #find the Encoded Packets
sequence = []                               #sequence list

NB_avg = 0

for i in range(5):
  generate_sequence(sequence_size, sequence)
  find_entropy(sequence)
  parsing_and_findEncodedPackets(dictionary_contents, encoded_packets)
  NB_avg = NB_avg + find_NB(dictionary_contents, encoded_packets)
  bits_per_phrase(encoded_packets)
  print("---------------------------------------------------------------------------\n")
  sequence.clear()

NB_avg = NB_avg / 5

print("NB_avg = ", NB_avg)
commp_ratio(NB_avg, sequence_size)
```
